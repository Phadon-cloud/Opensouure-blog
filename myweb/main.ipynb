```python
from fastapi import FastAPI, HTTPException, Request
from fastapi.responses import HTMLResponse
from fastapi.templating import Jinja2Templates 
import sqlite3
```

```python
from fastapi.staticfiles import StaticFiles


app = FastAPI()


app.mount("/static", StaticFiles(directory="static"), name="static")

templates = Jinja2Templates(directory="templates")

posts: list[dict] =   [
    {
        "id": 1,
        "author": "phadon.cloud",
        "title": "Why FastAPI Stands Out",
        "content": "FastAPI makes building APIs simple, efficient, and incredibly fast with minimal setup.",
        "date_posted": "January 30, 2026"
    },
    {
        "id": 2,
        "author": "phadon.cloud",
        "title": "Starting Your Journey with FastAPI",
        "content": "With clear documentation, clean routing, and modern design, FastAPI is easy to learn and enjoyable to use.",
        "date_posted": "January 30, 2026"
    }
]



@app.get("/", include_in_schema=False, name="home")
@app.get("/posts", include_in_schema=False, name="posts")
def home(request: Request):
    return templates.TemplateResponse(
        request,
        "home.html",
        {"posts": posts, "title": "Home"},
    )


@app.get("/api/posts")
def get_posts():
    return posts

```

```python
import asyncio
import uvicorn

config = uvicorn.Config(app)
server = uvicorn.Server(config)
await server.serve()
```
